```python
from IPython.display import clear_output

def printBoard(vector):
    print('')
    for i in range(5):
        if i % 2 != 0:
            print('-+-+-     -+-+-')
        else:
            for j in range(2):
                l = j + int(i/2) * 3
                print(vector[l] + '|', end = '')
            print(vector[l + 1] + '     ', end = '')
            for j in range(2):
                l = j + int(i/2) * 3
                print(legend[l] + '|', end = '')
            print(legend[l + 1])
    print('')

def userInput():
    while True:
        field = input(symbol + '\'s turn: ')
        if field == 'q':
            raise
        elif field not in legend:
            print('nope, choose available number from 1 - 9')
        else:
            break
    return int(field) - 1

def setSymbol(char):
    x = userInput()
    vector[x] = char
    legend[x] = ' '

def winCheck(vector):
    if ''.join(vector[0:3]) in check or \
       ''.join(vector[3:6]) in check or \
       ''.join(vector[6:9]) in check or \
       ''.join(vector[0::3]) in check or \
       ''.join(vector[1::3]) in check or \
       ''.join(vector[2::3]) in check or \
       ''.join(vector[0::4]) in check or \
       ''.join(vector[2::2]) in check:
        return True
    return False

vector = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
legend = ['1', '2', '3', '4', '5', '6', '7', '8', '9']
symbol = 'X'
check = ['XXX','OOO']

printBoard(vector)

for k in range(9):
    setSymbol(symbol)
    clear_output()
    printBoard(vector)
    if winCheck(vector) == True:
        print('player', symbol, 'won!')
        break
    if symbol == 'X':
        symbol = 'O'
    else:
        symbol = 'X'
else:
    print('it\'s a draw. boring.')
```
